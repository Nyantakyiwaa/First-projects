```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
data= pd.read_csv('linear_regression_data.csv')
data.head()
```

```python
#let us visualize the data
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.scatter(data.X, data.Y, color='blue',marker='+')
plt.show()
```

```python
#line of best fit
reg= linear_model.LinearRegression()
reg.fit(data[['X']],data.Y)
prediction=reg.predict([[98]])
print(prediction)
reg.coef_
reg.intercept_
```

```python
#let us plot the line of best fit
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.scatter(data.X, data.Y, color='blue',marker='+')
plt.plot(data.X,reg.predict(data[['X']]),color='red')
```
