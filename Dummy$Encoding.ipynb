```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
```

```python
#import the data and check it unquies
ds=pd.read_csv('employee_data.csv')
ds['Education'].nunique()
ds['Country'].nunique()
ds
```

```python
#using label encoder
le=LabelEncoder()
ds_le=ds
ds_le.Country=le.fit_transform(ds_le.Country)
ds_le
```

```python
#using pandas dummy for gender and education
ds_gen=pd.get_dummies(ds_le,columns=['Gender','Education'])
#merging the dummies and dataset
#using pandas dummy for gender and education
ds_gen=pd.get_dummies(ds_le,columns=['Gender','Education'])
# One-hot encode Gender and Education, drop first to avoid multicollinearity
ds_final = pd.get_dummies(ds_le, columns=['Gender', 'Education'], drop_first=True)
ds_final
```

```python
#defining X values and y as well
y=ds.Salary
X=ds_final.drop(['Salary','Name'],axis='columns').astype(int)
X
```

```python
#building regression model
model=LinearRegression()
model.fit(X,y)
model.score(X,y)

```

```python
#prediction
model.predict([[2,0,24,0,0,0]])
#the X variables in order as follow experience,country,age,gender_male, education_master and education_phd
```

```python
#let us train our data
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=10)
reg=LinearRegression()
reg.fit(X_train,y_train)
reg.score(X_test,y_test)
```
