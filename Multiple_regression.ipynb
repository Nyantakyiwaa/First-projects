```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
data=pd.read_csv('realistic_price_data.csv')
data.head()
```

```python
#line of best
reg=linear_model.LinearRegression()
reg.fit(data[['Size','Location_Score','Num_Features']],data.Price)
reg.predict(data[['Size','Location_Score','Num_Features']])
```
